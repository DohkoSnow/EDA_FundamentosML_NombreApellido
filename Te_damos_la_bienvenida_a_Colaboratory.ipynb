```
# Crear un histograma para la columna 'Age'
plt.hist(df['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```
# se calcula la asimetría de la columna 'Age'
skewness = df['Age'].skew()
print(f"Skewness of Age: {skewness}")
```

```
# se crea un diagrama de caja para la columna 'MonthlyIncome', se usa para visualizar la mediana, los cuartiles y
# los valores atipicos

sns.boxplot(x=df['MonthlyIncome'])
plt.xlabel('MonthlyIncome')
plt.title('Boxplot of MonthlyIncome')
plt.show()
```

```
# Calcular el IQR para la columna 'MonthlyIncome'
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

# Identificar valores atípicos
outliers = df[(df['MonthlyIncome'] < (Q1 - 1.5 * IQR)) | (df['MonthlyIncome'] > (Q3 + 1.5 * IQR))]
print(f"Number of outliers in MonthlyIncome: {len(outliers)}")
#los valores atipicos, suelen ser datos aislados.
#Los modelos son sensibles a estos datos
```

```
# Tabla de contingencia
contingency_table = pd.crosstab(df['Attrition'], df['JobSatisfaction'])

# Gráfico de barras
contingency_table.plot(kind='bar', stacked=True)
plt.xlabel('Attrition')
plt.ylabel('Frequency')
plt.title('Attrition vs JobSatisfaction')
plt.show()
```

```
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
plt.show()
```

Este análisis nos ayuda a entender por qué algunos empleados se van de la empresa y otros se quedan. Esta información es muy valiosa para la empresa, ya que le permite tomar decisiones para mejorar la retención de empleados y reducir la rotación.
